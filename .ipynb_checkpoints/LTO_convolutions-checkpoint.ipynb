```python
import hyperspy.api as hs
import matplotlib.pyplot as plt
LTO_data = hs.load("LTO_low_loss.msa").isig[:500]
LTO_data.set_microscope_parameters(beam_energy=300,convergence_angle=2,collection_angle=5)
LTO_data.add_elements(("Li","Ti","O"))
LTO_edge = LTO_data.isig[160:300]
LTO_edge=LTO_edge.remove_background((0,20))
LTO_edge.plot()
plt.show()
```
